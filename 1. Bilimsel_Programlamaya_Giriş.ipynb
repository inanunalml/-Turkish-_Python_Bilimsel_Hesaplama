## 1.6 Python'da Veri Tipleri

Python, programlama dünyasında verileri farklı şekillerde saklamak ve işlemek için çeşitli **veri tipleri** sunar. Veri tipleri, değişkenlerin bellekte nasıl depolanacağını ve hangi işlemlerin uygulanabileceğini belirler. Python'da veri tipleri, sayısal verilerden metinlere, mantıksal değerlerden karmaşık veri yapılarına kadar geniş bir yelpazeye sahiptir.

Bu bölümde, Python'da sıkça kullanılan veri tiplerini ve temel özelliklerini inceleyeceğiz.



Python, verileri farklı şekillerde saklamak ve işlemek için çeşitli veri tiplerini destekler.

### Sayısal Veri Tipleri
- **`int`**: Tam sayıları temsil eder.
  Örnek: `5`

- **`float`**: Ondalık sayıları temsil eder.
  Örnek: `3.14`

- **`complex`**: Karmaşık sayıları temsil eder.
  Örnek: `3 + 4j`

---

### Metin Veri Tipi
- **`str`**: Metin verilerini temsil eder.
  Örnek: `"Merhaba"`

---

### Mantıksal Veri Tipi
- **`bool`**: Mantıksal değerleri temsil eder (True veya False).
  Örnek: `True`

---

### Sıralı Veri Tipleri
- **`list`**: Sıralı ve değiştirilebilir koleksiyonları temsil eder.
  Örnek: `[1, 2, 3]`

- **`tuple`**: Sıralı ve değiştirilemez koleksiyonları temsil eder.
  Örnek: `(1, 2, 3)`

- **`range`**: Belirli bir aralıktaki sayıları temsil eder.
  Örnek: `range(5)`

---

### Haritalama Veri Tipi
- **`dict`**: Anahtar-değer çiftlerini saklayan koleksiyonları temsil eder.
  Örnek: `{"isim": "Ahmet", "yaş": 25}`

---

### Kümeler
- **`set`**: Sırasız, benzersiz ve değiştirilebilir koleksiyonları temsil eder.
  Örnek: `{1, 2, 3}`

- **`frozenset`**: Sırasız, benzersiz ve değiştirilemez koleksiyonları temsil eder.
  Örnek: `frozenset({1, 2, 3})`

---

### Binary Veri Tipleri
- **`bytes`**: Bayt dizilerini temsil eder.
  Örnek: `b"example"`

- **`bytearray`**: Değiştirilebilir bayt dizilerini temsil eder.
  Örnek: `bytearray(b"example")`

- **`memoryview`**: Bayt dizilerinin bellek görüntülerini temsil eder.
  Örnek: `memoryview(b"example")`


Bazı örnekleri inceleyelim. `type()` komutu ile verinin tipini görebiliriz.

## 1.7 Değişken Atama İşlemi

Bilimsel hesaplamalarda **değişken atama**, formüllerin, bağıntıların ve algoritmaların uygulanmasında kritik bir rol oynar. Değişkenler, verileri saklamak ve işlemek için kullanılır, böylece karmaşık hesaplamaları daha anlaşılır ve yönetilebilir hale getirir.


````python
x = "Tensör Akademi" #x isimli bir değişken tanımlandı ve bu değişkene "Değişken" kelimesi atandı.
````


Python'da değişken atama, `=` operatörü ile kolayca yapılır. Değişkenlere atanan değerler, sayısal veriler, metinler veya diğer veri tipleri olabilir. Aşağıdaki örneklerde, iki sayısal değişken tanımlanmakta ve bu değişkenler kullanılarak bazı basit işlemler gerçekleştirilmektedir.


````python
# Değişken atama
sayi1 = 10
sayi2 = 5

# Toplama işlemi
toplam = sayi1 + sayi2
print("Toplam:", toplam)
````

````python
# Çıkarma işlemi
fark = sayi1 - sayi2
print("Fark:", fark)
````

````python
# Çarpma işlemi
carpim = sayi1 * sayi2
print("Çarpım:", carpim)
````

````python
# Bölme işlemi
bolum = sayi1 / sayi2
print("Bölüm:", bolum)
````

````python
sayi1*sayi2
````

Hataya dikkat edelim. O halde bu işlemler farklı veri tipleri için değişik
anlamlara geliyor. Aşağıdaki tabloda bazı farklı veri tiplerinin işlem ilişkileri verilmiştir.

| İşlem / Veri Tipi           | `int`           | `float`             | `str`              | `list`             | `bool`                 |
| --------------------------- | --------------- | ------------------- | ------------------ | ------------------ | ---------------------- |
| `+` (toplama / birleştirme) | `3 + 5 → 8`     | `3.0 + 2 → 5.0`     | `'a' + 'b' → 'ab'` | `[1]+[2] → [1,2]`  | `True + False → 1`     |
| `-` (çıkarma)               | `5 - 2 → 3`     | `5.0 - 2 → 3.0`     | ❌ (Hata)           | ❌                  | `True - False → 1`     |
| `*` (çarpma / tekrar)       | `3 * 2 → 6`     | `2.5 * 2 → 5.0`     | `'a'*3 → 'aaa'`    | `[1]*3 → [1,1,1]`  | `True * 3 → 3`         |
| `/` (bölme)                 | `5 / 2 → 2.5`   | `5.0 / 2 → 2.5`     | ❌                  | ❌                  | `True / False → Hata`  |
| `//` (tam bölme)            | `5 // 2 → 2`    | `5.0 // 2 → 2.0`    | ❌                  | ❌                  | `True // False → Hata` |
| `%` (mod)                   | `5 % 2 → 1`     | `5.0 % 2 → 1.0`     | ❌                  | ❌                  | `True % False → Hata`  |
| `**` (üssü alma)            | `2 ** 3 → 8`    | `2.0 ** 3 → 8.0`    | ❌                  | ❌                  | `True ** 3 → 1`        |
| `len()`                     | ❌               | ❌                   | `'abc' → 3`        | `[1,2,3] → 3`      | ❌                      |
| `type()`                    | `type(5) → int` | `type(5.2) → float` | `type('a') → str`  | `type([1]) → list` | `type(True) → bool`    |



Sizler de farklı veri tipleri tanımlayarak bu işlemleri deneyiniz.

Bazı Örnekler



````python
# 🔹 Sayısal veri tipleri (int ve float)
a = 5
b = 2
x = 3.5
y = 1.2

print("int + int:", a + b)
print("int / int:", a / b)
print("int // int:", a // b)
print("float + float:", x + y)
print("int * float:", a * y)
````

````python
# 🔹 Metin (string) veri tipi
s1 = "Python"
s2 = " rocks!"

print("String birleştirme:", s1 + s2)
print("String tekrarlama:", s1 * 3)
# print("String çıkarma:", s1 - s2)  # ❌ Hata verir
````

````python
# 🔹 Liste (list) veri tipi
l1 = [1, 2, 3]
l2 = [4, 5]

print("Liste birleştirme:", l1 + l2)
print("Liste tekrarlama:", l1 * 2)
# print("Liste çıkarma:", l1 - l2)  # ❌ Hata verir
````

````python
# 🔹 Mantıksal (bool) veri tipi
p = True
q = False

print("Bool + Bool:", p + q)
print("Bool * 3:", p * 3)
print("Bool ile int:", p + 5)
# print("Bool / False:", p / q)  # ❌ ZeroDivisionError
````

## 1.8 Sonuç ve Öneriler

Python, dinamik veri tipi özelliği sayesinde, değişkenlerin veri tiplerini çalışma zamanı sırasında otomatik olarak belirler. Bu, Python'u esnek ve kullanıcı dostu bir dil haline getirir. Ancak, farklı veri tipleri ile yapılan aritmetik işlemlerde bazı kısıtlamalar ve hatalar ortaya çıkabilir.

- **Sayısal Veri Tipleri (`int`, `float`)**: Aritmetik işlemler genellikle sorunsuz çalışır. Bölme işlemi her zaman `float` sonucu verir.
- **Metin Veri Tipi (`str`)**: `+` operatörü metinleri birleştirirken, `*` operatörü metinleri tekrarlar. Diğer aritmetik işlemler (`-`, `/`) **TypeError** hatası verir.
- **Farklı Veri Tipleri (`int` ve `str`)**: Çoğu aritmetik işlem **TypeError** hatası verir. Ancak, `*` operatörü metni tekrarlar.

## Öneriler

1. **Veri Tipi Uyumluluğuna Dikkat Edin**: Farklı veri tipleri ile aritmetik işlemler yaparken, veri tiplerinin uyumlu olmasına dikkat edin. Gerekirse veri tiplerini dönüştürün (örneğin, `int()` veya `float()` fonksiyonları ile).

2. **Hata Yönetimi**: Farklı veri tipleri ile yapılan işlemlerde oluşabilecek **TypeError** hatalarını yönetmek için `try-except` bloklarını kullanın.

3. **Kodun Okunabilirliği**: Değişkenlere anlamlı isimler vererek ve yorum satırları ekleyerek kodun okunabilirliğini artırın.

4. **Test ve Doğrulama**: Yazdığınız kodları farklı veri tipleri ile test ederek, beklenmeyen hataların önüne geçin.

Python'ın dinamik veri tipi özelliği, hızlı geliştirme ve esneklik sağlar. Ancak, veri tiplerinin uyumluluğuna dikkat edilmesi, hataların önlenmesi ve kodun güvenilirliği açısından kritik öneme sahiptir.
